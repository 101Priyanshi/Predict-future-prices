#  PREPARING THE DATA BEFORE APPLYING FACEBOOK PROPHET TOOL 

```python
dataset#we just need the date and the avg price
```

```python
dataset_fb=dataset[['Date','AveragePrice']]
```

```python
dataset_fb
```

```python
dataset_fb=dataset_fb.rename(columns={'Date':'ds','AveragePrice':'y'})#that's how prophet req its data to be
```

```python
dataset_fb
```

# TRAINING THE WHOLE MODEL ON THE DATASET AND MAKING PREDICTIONS 

```python
m=Prophet()
m.fit(dataset_fb)
```

```python
# Forcasting into the future
future = m.make_future_dataframe(periods=365)#if we want to predict one year ahead
forecast = m.predict(future)
```

```python
forecast#fbprophet not only tells the trend but also the confidence boundary i.e. how much lower it will get (yhat_lower) etc
```

```python
figure = m.plot(forecast,xlabel='Date',ylabel='Price')
```

```python
figure2 = m.plot_components(forecast)
```

# TRAINING THE WHOLE MODEL ON THE DATASET AND MAKING PREDICTIONS (REGION SPECIFIC) 

```python
avocado_df = pd.read_csv('avocado.csv')# dataframes creation for both training and testing datasets 
```

```python
avocado_df_sample=avocado_df[avocado_df['region']=='West']# Select specific region
```

```python
avocado_df_sample = avocado_df_sample.sort_values('Date') 
```

```python
plt.plot(avocado_df_sample['Date'],avocado_df_sample['AveragePrice'])
```

```python
avocado_df_sample = avocado_df_sample.rename(columns = {'Date':'ds','AveragePrice':'y'})
```

```python
m = Prophet()
m.fit(avocado_df_sample)
# Forcasting into the future
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```

```python
figure = m.plot(forecast, xlabel='Date', ylabel='Price')
```

```python
figure3 = m.plot_components(forecast)
```
